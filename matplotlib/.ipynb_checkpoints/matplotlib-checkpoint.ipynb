#### Basic Graph

```python
x = [1, 2, 3]
y = [1, 2, 3]
# y = [2, 4, 6]

plt.plot(x, y, label='1:1',
        color='green',
        linestyle='dashed',
        marker='o', ## marks the actual cross point of (x, y)
        )

nums = np.arange(1, 10, 0.1)
plt.plot(nums, np.sin(nums), label='sin')

plt.title('First pyplot', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})

plt.xlabel('Time', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 16})
plt.ylabel('Distance', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 16})

# plt.xticks(list(range(1, 4)))
# plt.yticks(list(range(1, 4)))


plt.legend()  # requires lebel to be present

plt.show()
```

### Bar chart

```python
labels = ['A', 'B', 'C']
values = [1, 4, 2]

bars = plt.bar(labels, values)

bars[0].set_hatch('/')
bars[1].set_hatch('o')

plt.figure(figsize=(8, 4))

plt.show()
```
